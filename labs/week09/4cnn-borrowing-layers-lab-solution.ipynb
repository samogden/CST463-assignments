```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.applications.vgg16 import VGG16
from tensorflow.keras.applications.imagenet_utils import decode_predictions
from tensorflow.keras.utils import plot_model
from tensorflow.keras.datasets import cifar10
import tensorflow as tf
```

```python
model = VGG16()
```

```python
model.summary()
```

```python
plot_model(model)
```

```python
(X_train, y_train), (X_test, y_test) = cifar10.load_data()
```

```python
print(X_train.shape)
print(y_train.shape)
```

```python
idx = np.random.choice(X_train.shape[0], 100)
X_sample = X_train[idx]
y_sample = y_train[idx]
```

```python
# CIFAR 10 labels
print(y_sample[:10])
```

```python
plt.imshow(X_sample[0]);
```

```python
plt.imshow(X_sample[1])
```

```python
X_sample_resized = tf.image.resize(X_sample, (224,224), antialias=True)
```

```python
X_sample_resized = np.array(X_sample_resized)/255.0
```

```python
plt.imshow(X_sample_resized[0]);
```

```python
y_pred_sample = model.predict(X_sample_resized)
```

```python
y_pred_sample.shape
```
