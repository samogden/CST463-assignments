## Gradient boosting lab

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.datasets import load_diabetes
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error
```

#### Load the data

```python
# load the diabetes data
# It may have been faster to load X,y directly, but I wanted to see the variable names.
df = load_diabetes(as_frame=True)['frame']
```

```python
df.info()
```

#### Create training and test sets

```python
target = 'target'
predictors = list(df.columns)
predictors.remove(target)

X = df[predictors].values
y = df[target].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

#### Use a simple classification tree

```python
tree = DecisionTreeRegressor()
rmse_tree = -cross_val_score(tree, X_train, y_train, cv=5, scoring='neg_root_mean_squared_error').mean()
print(f'baseline RMSE for decision tree: {rmse_tree:.2f}')
```
